```python
import pandas as pd
import numpy as np
import statsmodels.api as sm
import statsmodels.formula.api as smf
```

```python
ich_df = pd.read_csv('ich_df.csv', index_col=0)
```

```python
ich_df.head()
```

```python
ich_df.sort_values(by=['PM2.5'], ascending=False)
```

```python
ich_df.sort_values(by=['PM10'], ascending=False)
```

```python
ich_df = ich_df.drop('geometry', axis=1)
```

```python
gdf_2020 = pd.read_csv('gdf_2020.csv')
```

```python
total_df = pd.merge(gdf_2020,ich_df, how='inner', on='법정동')
```

```python
total_df.head()
```

```python
total_df.columns
```

```python
model1 = smf.ols(formula = 'PM10 ~ 제조업_기업수 + 건축착공_신고수 + 공동주택수 + 환경오염물질배출_사업소수 + 사업장폐기물배출_사업소수', data = total_df)
result1 = model1.fit()
result1.summary()
```
