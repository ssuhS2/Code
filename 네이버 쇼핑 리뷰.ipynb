```python
def move_next(driver):
    #page_bar=driver.find_elements_by_css_selector('#section_review > div.pagination_pagination__2M9a4')[0]
    #pages=page_bar.find_elements_by_css_selector('a')
    #del pages[0]
    #page_now=page_bar.find_elements_by_css_selector('#section_review > div.pagination_pagination__2M9a4 > a.pagination_now__gZWGP')[0].text.replace('현재 페이지','').strip()
    for page in pages:
        del pages[0]
        page_num=page.text.strip()
        if page_num in ['맨앞','이전','맨뒤']:
            pass
        elif page_num =='다음':
            page.send_keys('\n')
            driver.implicitly_wait(3)
            time.sleep(2+random.uniform(0,1))
            return False
        elif int(page_num)+1 >int(page_now):
            page.send_keys('\n')
            driver.implicitly_wait(3)
            time.sleep(1+random.uniform(0,1))
            return False
        return True
```
