```python
workbook = openpyxl.load_workbook('C:/Users/CPBUserN/Desktop/KTmhows/빅카인즈_191016~201016_본문.xlsx')
worksheet = workbook['sheet']
```

```python
list_var = []
cell_range=worksheet['A2':'A232']

for row in cell_range:
    for cell in row:
        list_var.append(cell.value)
```

```python
twitter = Twitter() 
morphs = []
for sentence in list_var: 
    morphs.append(twitter.pos(sentence))
```

```python
noun_adj_adv_list=[] 

for sentence in morphs : 
    for word, tag in sentence : 
        if tag in ['Noun']and ("인기" not in word)and ("점" not in word)and ("중심" not in word)and ("경제" not in word)and ("그룹" not in word)and ("업계" not in word)and ("인기" not in word)and ("대표" not in word)and ("올해" not in word)and ("출시" not in word)and ("로" not in word)and ("층" not in word)and ("위" not in word)and ("롯데" not in word)and ("아테" not in word)and ("지난" not in word)and ("최근" not in word)and ("온" not in word)and ("를" not in word)and ("의" not in word)and ("시장" not in word) and ("기자" not in word)and ("트렌드" not in word)and ("것" not in word)and ("세대" not in word)and ("이" not in word) and ("등" not in word)and ("내" not in word)and ("나" not in word)and ("수"not in word) and("게"not in word)and("말"not in word)and ("때" not in word)and ("진짜" not in word): 
            noun_adj_adv_list.append(word) 
```
