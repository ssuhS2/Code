```python
list_var = []
cell_range=worksheet['A2':'A561']

for row in cell_range:
    for cell in row:
        list_var.append(cell.value)
```

```python
twitter = Twitter() 
morphs = []
for sentence in list_var: 
    morphs.append(twitter.pos(sentence))
    
print(morphs)
```

```python
noun_adj_adv_list=[] 

for sentence in morphs : 
    for word, tag in sentence : 
        if tag in ['Noun'] and ("ê²ƒ" not in word) and ("ë‚´" not in word)and ("ë‚˜" not in word)and ("ìˆ˜"not in word) and("ê²Œ"not in word)and("ë§"not in word)and ("ë•Œ" not in word)and ("ì§„ì§œ" not in word): 
            noun_adj_adv_list.append(word) 
            
print(noun_adj_adv_list)
```
