```python
from konlpy.tag import Twitter 
from collections import Counter
```

```python
import openpyxl
```

```python
workbook = openpyxl.load_workbook('C:/Users/CPBUserN/Desktop/KTmhows/유튜브크롤링_90년대생모여라.xlsx')
worksheet = workbook['Sheet1']
```

```python
list_var = []
cell_range=worksheet['A2':'A561']

for row in cell_range:
    for cell in row:
        list_var.append(cell.value)
```

```python
twitter = Twitter() 
morphs = []
for sentence in list_var: 
    morphs.append(twitter.pos(sentence))
    
print(morphs)
```

```python
noun_adj_adv_list=[] 

for sentence in morphs : 
    for word, tag in sentence : 
        if tag in ['Noun'] and ("것" not in word) and ("내" not in word)and ("나" not in word)and ("수"not in word) and("게"not in word)and("말"not in word)and ("때" not in word)and ("진짜" not in word): 
            noun_adj_adv_list.append(word) 
            
print(noun_adj_adv_list)
```

```python
count = Counter(noun_adj_adv_list)
words = dict(count.most_common())
words
```

```python
from wordcloud import WordCloud 

import matplotlib.pyplot as plt

import nltk
from nltk.corpus import stopwords
%matplotlib inline

import matplotlib
from IPython.display import set_matplotlib_formats
matplotlib.rc('font',family = 'Malgun Gothic')

set_matplotlib_formats('retina')

matplotlib.rc('axes',unicode_minus = False)
```

```python
from matplotlib import font_manager, rc
font_fname ="C:/Windows/Fonts/H2GTRE.TTF"    # A font of your choice
font_name = font_manager.FontProperties(fname=font_fname).get_name()
rc('font', family=font_name)
```

```python
from matplotlib.colors import LinearSegmentedColormap
colors = ["#c71f2e", "#11584a", "#FF8B8B", "#384883", "#f8a39e", "#d7942c"]
cmap = LinearSegmentedColormap.from_list("mycmap", colors)
wordcloud = WordCloud(font_path = font_fname, background_color='white',colormap = cmap, width=1500, height=1000,  stopwords=stopwords).generate_from_frequencies(words) 
plt.imshow(wordcloud) 
plt.axis('off') 
plt.show()
```
