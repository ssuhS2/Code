```python
from konlpy.tag import Twitter 
from collections import Counter
import openpyxl
```

```python
workbook = openpyxl.load_workbook('C:/Users/CPBUserN/Desktop/KTmhows/빅카인즈_191016~201016.xlsx')
worksheet = workbook['sheet']
```

```python
list_var = []
cell_range=worksheet['A2':'AJ232']

for row in cell_range:
    for cell in row:
        list_var.append(cell.value)

print(list_var)
```

```python
noun_adj_adv_list=[] 

for sentence in list_var : 
    if (sentence != None)and (sentence != "차별화") and (sentence != "차별화")and (sentence != "부동산")and (sentence != "언택트")  and (sentence != "론칭") and (sentence != "코로나19")and (sentence != "온라인")and (sentence != "서울")  and (sentence != "한국")  and (sentence != "40대")  and (sentence != "소비자")  and (sentence != "고객들") and (sentence != "만큼")and (sentence != "경기도") and (sentence != "1위") and (sentence != "30대")and (sentence != "20대")and (sentence != "3040세대") and (sentence != "관계자"): 
        noun_adj_adv_list.append(sentence) 
            
print(noun_adj_adv_list)
```

```python
count = Counter(noun_adj_adv_list)
words = dict(count.most_common())
words["뉴트로"]=words.pop('뉴트')
words
```

```python
from wordcloud import WordCloud 

import matplotlib.pyplot as plt

import nltk
from nltk.corpus import stopwords
%matplotlib inline

import matplotlib
from IPython.display import set_matplotlib_formats
matplotlib.rc('font',family = 'Malgun Gothic')

set_matplotlib_formats('retina')

matplotlib.rc('axes',unicode_minus = False)
```

```python
from matplotlib import font_manager, rc
font_fname ="C:/Windows/Fonts/H2GTRE.TTF"    # A font of your choice
font_name = font_manager.FontProperties(fname=font_fname).get_name()
rc('font', family=font_name)
```

```python
from matplotlib.colors import LinearSegmentedColormap
colors = ["#c71f2e", "#11584a", "#FF8B8B", "#384883", "#f8a39e", "#d7942c"]
cmap = LinearSegmentedColormap.from_list("mycmap", colors)
wordcloud = WordCloud(font_path = font_fname, background_color='white',colormap = cmap, width=1500, height=1000,  stopwords=stopwords).generate_from_frequencies(words) 
plt.imshow(wordcloud) 
plt.axis('off') 
plt.show()
```
