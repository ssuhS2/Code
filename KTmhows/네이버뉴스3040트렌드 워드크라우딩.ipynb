```python
from konlpy.tag import Twitter 
from collections import Counter
```

```python
import openpyxl
workbook = openpyxl.load_workbook('C:/Users/CPBUserN/Desktop/KTmhows/3040트렌드190101~191231.xlsx')
worksheet = workbook['sheet1']
```

```python
list_var = []
cell_range=worksheet['A2':'B1009']

for row in cell_range:
    for cell in row:
        list_var.append(cell.value)
```

```python
twitter = Twitter() 
morphs = []
for sentence in list_var: 
    morphs.append(twitter.pos(sentence))
```

```python
noun_adj_adv_list=[] 

for sentence in morphs : 
    for word, tag in sentence : 
        if tag in ['Noun']and ("남성" not in word)and ("여성" not in word)and ("올해" not in word)and ("신한은행" not in word) and ("한국" not in word)and ("코리아" not in word)and ("반영" not in word)and ("고객" not in word)and ("롯데" not in word)and ("보기" not in word)and ("번가" not in word)and ("최근" not in word)and ("중" not in word)and ("전체" not in word)and ("관련" not in word)and ("트렌드" not in word)and ("세대" not in word)and ("제" not in word) and ("명" not in word)and ("것" not in word)and ("점" not in word)and ("로" not in word)and ("건" not in word) and ("후" not in word)and ("이" not in word)and ("위" not in word) and ("층" not in word)  and ("등" not in word)  and ("를" not in word) and ("배" not in word)and ("나" not in word)and ("수"not in word) and("게"not in word)and("말"not in word)and ("때" not in word)and ("진짜" not in word): 
            noun_adj_adv_list.append(word) 
```

```python
count = Counter(noun_adj_adv_list)
words = dict(count.most_common())
words
```

```python
from wordcloud import WordCloud 

import matplotlib.pyplot as plt

import nltk
from nltk.corpus import stopwords
%matplotlib inline

import matplotlib
from IPython.display import set_matplotlib_formats
matplotlib.rc('font',family = 'Malgun Gothic')

set_matplotlib_formats('retina')

matplotlib.rc('axes',unicode_minus = False)
```

```python
from matplotlib import font_manager, rc
font_fname ="C:/Windows/Fonts/H2GTRE.TTF"    # A font of your choice
font_name = font_manager.FontProperties(fname=font_fname).get_name()
rc('font', family=font_name)
```

```python
from matplotlib.colors import LinearSegmentedColormap
colors = ["#c71f2e", "#11584a", "#FF8B8B", "#384883", "#f8a39e", "#d7942c"]
cmap = LinearSegmentedColormap.from_list("mycmap", colors)
wordcloud = WordCloud(font_path = font_fname, background_color='white',colormap = cmap, width=1500, height=1000,  stopwords=stopwords).generate_from_frequencies(words) 
plt.imshow(wordcloud) 
plt.axis('off') 
plt.show()
```
